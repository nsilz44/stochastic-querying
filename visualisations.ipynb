```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_CDF_expected(approxRatios,heuristicRatios,title,xlabel,ylabel):
    approxX = np.sort(approxRatios)
    approxY = 1 * np.arange(len(approxRatios)) / (len(approxRatios) - 1)    
    plt.plot(approxX, approxY)
    heuristicX = np.sort(heuristicRatios)
    heuristicY = 1 * np.arange(len(heuristicRatios)) / (len(heuristicRatios) - 1)    
    plt.plot(heuristicX,heuristicY)
    plt.show()
```

```python
intervalsMinDf = pd.read_csv('intervalsTests.csv')
intervalsMinDf.head(12)
```

```python
approxRatios = intervalsMinDf.iloc[4]
approxRatios = [float(idx) for idx in approxRatios[-2][1:-1].split(', ')]
heuristicRatios = intervalsMinDf.iloc[4]
heuristicRatios = [float(idx) for idx in heuristicRatios[-1][1:-1].split(', ')]
plot_CDF_expected(approxRatios,heuristicRatios,'title','l','y')
```
