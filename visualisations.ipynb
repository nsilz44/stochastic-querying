```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def plot_CDF_expected(approxRatios,heuristicRatios,title,xlabel,ylabel):
    approxX = np.sort(approxRatios)
    approxY = 1 * np.arange(len(approxRatios)) / (len(approxRatios) - 1)    
    plt.plot(approxX, approxY,color='cyan', label = 'Algorithm 1')
    heuristicX = np.sort(heuristicRatios)
    heuristicY = 1 * np.arange(len(heuristicRatios)) / (len(heuristicRatios) - 1)    
    plt.plot(heuristicX,heuristicY,color='purple', label = 'Algorithm 2')
    plt.legend(loc='lower right')
    plt.xlabel('Competitive ratio')
    plt.ylabel('Culmutive distribution')
    plt.show()

def plotAverage_expected(df,title):
    ax1 = df.plot(kind='scatter', x ='Test',y='Expected competitive ratio of approximation algorithm',color='cyan',label='Algorithm 1',marker="+")
    ax2 = df.plot(kind='scatter', x ='Test',y='Expected competitive ratio of heuristic approximation algorithm',color='purple', label='Algorithm 2',marker="x",ax=ax1)
    ax1.set_ylabel('Expected competitive ratio')
    plt.show()

def plot_ratios(df,title):
    approxUni = []
    heuristicUni = []
    approxCost = []
    heuristicCost = []
    testArray = []
    for i in range(len(df)):
        approxArray = [float(idx) for idx in df.iloc[i][-2][1:-1].split(', ')]
        heuristicArray = [float(idx) for idx in df.iloc[i][-1][1:-1].split(', ')]
        approxUni.append(approxArray.count(1.0)/ df.iloc[i][-3])
        approxArray = [i for i in approxArray if i != 1]
        approxCost.append(np.average(approxArray))
        heuristicUni.append(heuristicArray.count(1.0)/ df.iloc[i][-3])
        heuristicArray = [i for i in heuristicArray if i != 1]
        heuristicCost.append(np.average(heuristicArray))
        testArray.append(df.iloc[i][1])
    print('test',testArray)
    print('approxUni',approxUni)
    print('heuristicUni',heuristicUni)
    print('approxCost',approxCost)
    print('heuristicCost',heuristicCost)

    
    
    xaxis = np.arange(len(testArray)) 
    plt.bar(xaxis - 0.2, approxUni, 0.4, label = 'Algorithm 1',color='cyan')
    plt.bar(xaxis + 0.2, heuristicUni, 0.4, label = 'Algorithm 2',color='purple')
    plt.xticks(xaxis, testArray)
    plt.xlabel("Tests")
    plt.ylabel("Optimum Ratio")
    plt.legend(loc='lower right')
    plt.show()

    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.scatter(x =testArray,y=approxCost,color='cyan',label='Algorithm 1',marker="+")
    ax1.scatter(x =testArray,y=heuristicCost,color='purple', label='Algorithm 2',marker="x")
    ax1.set_ylabel('Expected Cost ratio')
    plt.legend()
    plt.show()

   
```

```python
uniformMindf = pd.read_csv('uniformityTests.csv')
plotAverage_expected(uniformMindf,'LOl')
plot_ratios(uniformMindf,'t')
uniformMindf.head(12)
```

```python
probMinDf = pd.read_csv('probabilityTests.csv')
plotAverage_expected(probMinDf,'LOl')
plot_ratios(probMinDf,'t')
probMinDf.head(12)
```

HYPERGRAPH VISUALISATIONS

```python
def plot_CDF_Hyper_expected(approxRatios,heuristicRatios,title,xlabel,ylabel):
    approxX = np.sort(approxRatios)
    approxY = 1 * np.arange(len(approxRatios)) / (len(approxRatios) - 1)    
    plt.plot(approxX, approxY,color='g', label = 'Algorithm 4')
    heuristicX = np.sort(heuristicRatios)
    heuristicY = 1 * np.arange(len(heuristicRatios)) / (len(heuristicRatios) - 1)    
    plt.plot(heuristicX,heuristicY,color='orange', label = 'Algorithm 5')
    plt.legend(loc='lower right')
    plt.xlabel('Competitive ratio')
    plt.ylabel('Culmutive distribution')
    plt.show()

def plotAverage_comp(df,title):
    ax1 = df.plot(kind='scatter', x ='Test',y='Competitive ratio of threshold algorithm',color='g',label='Algorithm 4',marker="+")
    ax2 = df.plot(kind='scatter', x ='Test',y='Competitive ratio of bestVc algorithm',color='orange', label='Algorithm 5',marker="x",ax=ax1)
    ax1.set_ylabel('Competitive ratio')
    plt.show()

def plot_Hyper_ratios(df,title):
    thresholdUni = []
    bestVCUni = []
    thresholdCost = []
    bestVcCost = []
    testArray = []
    for i in range(len(df)):
        thresholdArray = [float(idx) for idx in df.iloc[i][-2][1:-1].split(', ')]
        bestVCArray = [float(idx) for idx in df.iloc[i][-1][1:-1].split(', ')]
        thresholdUni.append(thresholdArray.count(1.0)/ df.iloc[i][-3])
        thresholdArray = [i for i in thresholdArray if i != 1]
        thresholdCost.append(np.average(thresholdArray))
        bestVCUni.append(bestVCArray.count(1.0)/ df.iloc[i][-3])
        bestVCArray = [i for i in bestVCArray if i != 1]
        bestVcCost.append(np.average(bestVCArray))
        testArray.append(df.iloc[i][1])
    print('test',testArray)
    print('thresholdUni',thresholdUni)
    print('bestVCUni',bestVCUni)
    print('thresholdCost',thresholdCost)
    print('bestVcCost',bestVcCost)
    
    xaxis = np.arange(len(testArray)) 
    plt.bar(xaxis - 0.2, thresholdUni, 0.4, label = 'Algorithm 4',color='g')
    plt.bar(xaxis + 0.2, bestVCUni, 0.4, label = 'Algorithm 5',color='orange')
    plt.xticks(xaxis, testArray)
    plt.xlabel("Tests")
    plt.ylabel("Optimum Ratio")
    plt.legend(loc='lower right')
    plt.show()
    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    ax1.scatter(x =testArray,y=thresholdCost,label='Algorithm 4',marker="+",color='g')
    ax1.scatter(x =testArray,y=bestVcCost, label='Algorithm 5',marker="x",color='orange')
    ax1.set_ylabel('Expected Cost ratio')
    plt.show()
```

```python
from scipy.stats import uniform
from scipy.stats import beta
from scipy.stats import semicircular
from scipy.stats import gausshyper
from scipy.stats import rdist
from scipy.stats import bradford


```

```python
mean, var, skew, kurt = uniform.stats(moments='mvsk')
fig, ax = plt.subplots(1, 1)
x = np.linspace(uniform.ppf(0.01),uniform.ppf(0.99), 100)
ax.plot(x, uniform.pdf(x),'r-', lw=5, alpha=0.6, label='uniform pdf')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
betaA, betaB = 2, 0.8
gaussA,gaussB,gaussC,gaussZ = 10, 3, 2, 5
rDistC = 1.6
bradfordC = 15
mean, var, skew, kurt = beta.stats(betaA, betaB, moments='mvsk')
fig, ax = plt.subplots(1, 1)
x = np.linspace(beta.ppf(0.01, betaA, betaB),beta.ppf(0.99, betaA, betaB), 100)
ax.plot(x, beta.pdf(x, betaA, betaB),'r-', lw=5, alpha=0.6, label='beta pdf')
plt.xticks([]),plt.yticks([])
plt.show()
```

```python
fig, ax = plt.subplots(1, 1)
mean, var, skew, kurt = gausshyper.stats(gaussA, gaussB, gaussC, gaussZ, moments='mvsk')
x = np.linspace(gausshyper.ppf(0.01, gaussA, gaussB, gaussC, gaussZ),gausshyper.ppf(0.99, gaussA, gaussB, gaussC, gaussZ), 100)
ax.plot(x, gausshyper.pdf(x, gaussA, gaussB, gaussC, gaussZ),'r-', lw=5, alpha=0.6, label='gausshyper pdf')
plt.xticks([]),plt.yticks([])
plt.show()
```
